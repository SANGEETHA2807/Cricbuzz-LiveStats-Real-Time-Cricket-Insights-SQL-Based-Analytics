```python
import psycopg2
from psycopg2 import sql
import requests
from datetime import datetime


# ----------------------------
# Config - update these
# ----------------------------
# DB_USER = "postgres"
# DB_PASSWORD = "hello0!MSL"
# DB_HOST = "localhost"
# DB_PORT = "5433"
# DB_NAME = "cricbuzz"

API_KEY = "7ee193bd09mshd975546ea0e50bcp1756fbjsnb79a9ed72db8"
API_HOST = "cricbuzz-cricket.p.rapidapi.com"
HEADERS = {"x-rapidapi-key": API_KEY, "x-rapidapi-host": API_HOST}

# ----------------------------
# Connect to PostgreSQL
# ----------------------------
conn = psycopg2.connect(
    dbname="cricbuzz",  # Default database to connect
    user="postgres",
    password="hello01!MSL",
    host="localhost",
    port="5433"
)
cur = conn.cursor()

def run(query, params=None):
    cur.execute(query, params or ())
    conn.commit()

# ----------------------------
# 1️⃣ Teams Table
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS Teams (
    team_id INTEGER PRIMARY KEY,
    team_name TEXT,
    teamSName TEXT,
    country TEXT
)
""")

# fetch teams
# teams_url = "https://cricbuzz-cricket.p.rapidapi.com/teams/v1/international"
# teams_response = requests.get(teams_url, headers=HEADERS).json()
# teams_list = []

# for team in teams_response.get("list", []):
#     if "teamId" in team:
#         teams_list.append({
#             "id": team.get("teamId"),
#             "name": team.get("teamName"),
#             "sName": team.get("teamSName"),
#             "country": team.get("countryName")
#         })
# for t in teams_list:
#     if t["id"] is None:
#         continue  # skip invalid entries
#     run("""
#     INSERT INTO teams (team_id, team_name, teamsname, country)
#     VALUES (%s, %s, %s, %s)
#     ON CONFLICT (team_id) DO UPDATE
#     SET team_name = EXCLUDED.team_name,
#         teamsname = EXCLUDED.teamsname,
#         country = EXCLUDED.country
#     """, (int(t["id"]), t["name"], t["sName"], t["country"]))

# # for t in teams_list:
# #     run("""
# #     INSERT INTO teams (team_id, team_name, teamSName, country)
# #     VALUES (?, ?, ?, ?)
# #     """, (t["id"], t["name"], t["sName"], t["country"]))
    
# ----------------------------
# 2️⃣ PlayersQuestion1 (single-team example)
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS PlayersQuestion1 (
    player_id INTEGER PRIMARY KEY,
    team_id INTEGER REFERENCES Teams(team_id),
    full_name TEXT,
    batting_style TEXT,
    bowling_style TEXT,
    role TEXT
)
""")

# example: team_id = 2 (you can loop multiple ids if needed)
team_id = 2
url = f"https://cricbuzz-cricket.p.rapidapi.com/teams/v1/{team_id}/players"
data = requests.get(url, headers=HEADERS).json()

playersquestion1 = []
current_role = None
for player in data.get("player", []):
    if "id" not in player:
        current_role = player.get("name")
        continue
    playersquestion1.append({
        "id": player.get("id"),
        "team_id": team_id,
        "name": player.get("name"),
        "batting": player.get("battingStyle"),
        "bowling": player.get("bowlingStyle"),
        "role": current_role
    })

for p in playersquestion1:
    run("""
    INSERT INTO PlayersQuestion1 (player_id, team_id, full_name, batting_style, bowling_style, role)
    VALUES (%s, %s, %s, %s, %s, %s)
    ON CONFLICT (player_id) DO UPDATE
    SET team_id = EXCLUDED.team_id,
        full_name = EXCLUDED.full_name,
        batting_style = EXCLUDED.batting_style,
        bowling_style = EXCLUDED.bowling_style,
        role = EXCLUDED.role
    """, (p["id"], p["team_id"], p["name"], p["batting"], p["bowling"], p["role"]))

# ----------------------------
# 3️⃣ Matches Table
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS Matches (
    match_id INTEGER PRIMARY KEY,
    series_id INTEGER,
    series_name TEXT,
    team1id INTEGER,
    team2id INTEGER,
    team1 TEXT,
    team2 TEXT,
    venue_id INTEGER,
    venuename TEXT,
    city TEXT,
    format TEXT,
    status TEXT,
    state TEXT,
    start_time TEXT,
    result TEXT,
    winner_team_id INTEGER,
    victory_margin TEXT,
    victory_type TEXT,
    toss_winner_id INTEGER,
    toss_decision TEXT
)
""")

match_urls = [
    "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/live",
    "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/recent"
]

matches_list = []
for url in match_urls:
    data = requests.get(url, headers=HEADERS).json()
    for type_match in data.get("typeMatches", []):
        for series in type_match.get("seriesMatches", []):
            series_data = series.get("seriesAdWrapper", {})
            series_name = series_data.get("seriesName")
            series_id = series_data.get("seriesId")
            for match in series_data.get("matches", []):
                info = match.get("matchInfo", {})
                team1 = info.get("team1", {}).get("teamName")
                team2 = info.get("team2", {}).get("teamName")
                team1id = info.get("team1", {}).get("teamId")
                team2id = info.get("team2", {}).get("teamId")
                toss_winner_name = info.get("toss", {}).get("winnerTeam", {}).get("teamName")
                toss_decision = info.get("toss", {}).get("decision")
                toss_winner_id = team1id if toss_winner_name == team1 else (team2id if toss_winner_name == team2 else None)
                status = info.get("status")
                winner_id = None
                victory_margin = None
                victory_type = None
                if status and "won by" in status:
                    parts = status.split(" won by ")
                    winner_name = parts[0].strip()
                    details = parts[1].split(" ")
                    victory_margin = details[0]
                    victory_type = details[1] if len(details) > 1 else ""
                    if winner_name == team1:
                        winner_id = team1id
                    elif winner_name == team2:
                        winner_id = team2id
                matches_list.append({
                    "match_id": info.get("matchId"),
                    "series_id": series_id,
                    "series_name": series_name,
                    "team1id": team1id,
                    "team2id": team2id,
                    "team1": team1,
                    "team2": team2,
                    "venue_id": info.get("venueInfo", {}).get("id"),
                    "venuename": info.get("venueInfo", {}).get("ground"),
                    "city": info.get("venueInfo", {}).get("city"),
                    "format": info.get("matchFormat"),
                    "status": status,
                    "state": info.get("state"),
                    "start_time": info.get("startDate"),
                    "result": info.get("result"),
                    "winner_team_id": winner_id,
                    "victory_margin": victory_margin,
                    "victory_type": victory_type,
                    "toss_winner_id": toss_winner_id,
                    "toss_decision": toss_decision
                })

for m in matches_list:
    run("""
    INSERT INTO Matches
    (match_id, series_id, series_name, team1id, team2id, team1, team2, venue_id, venuename, city, format,
     status, state, start_time, result, winner_team_id, victory_margin, victory_type, toss_winner_id, toss_decision)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    ON CONFLICT (match_id) DO UPDATE
    SET series_id = EXCLUDED.series_id,
        series_name = EXCLUDED.series_name,
        team1id = EXCLUDED.team1id,
        team2id = EXCLUDED.team2id,
        team1 = EXCLUDED.team1,
        team2 = EXCLUDED.team2,
        venue_id = EXCLUDED.venue_id,
        venuename = EXCLUDED.venuename,
        city = EXCLUDED.city,
        format = EXCLUDED.format,
        status = EXCLUDED.status,
        state = EXCLUDED.state,
        start_time = EXCLUDED.start_time,
        result = EXCLUDED.result,
        winner_team_id = EXCLUDED.winner_team_id,
        victory_margin = EXCLUDED.victory_margin,
        victory_type = EXCLUDED.victory_type,
        toss_winner_id = EXCLUDED.toss_winner_id,
        toss_decision = EXCLUDED.toss_decision
    """, (
        m["match_id"], m["series_id"], m["series_name"], m["team1id"], m["team2id"], m["team1"], m["team2"],
        m["venue_id"], m["venuename"], m["city"], m["format"], m["status"], m["state"], m["start_time"],
        m["result"], m["winner_team_id"], m["victory_margin"], m["victory_type"], m["toss_winner_id"], m["toss_decision"]
    ))


team_ids_in_matches = set()
for m in matches_list:
    if m["team1id"]:
        team_ids_in_matches.add(m["team1id"])
    if m["team2id"]:
        team_ids_in_matches.add(m["team2id"])

team_ids_in_matches = {}
for m in matches_list:
    # Add team1
    if m.get("team1id") and m.get("team1name"):
        team_ids_in_matches[m["team1id"]] = {
            "name": m["team1name"],
            "sName": m.get("team1sname", ""),
            "country": None  # leave empty for now
        }
    # Add team2
    if m.get("team2id") and m.get("team2name"):
        team_ids_in_matches[m["team2id"]] = {
            "name": m["team2name"],
            "sName": m.get("team2sname", ""),
            "country": None
        }

# Insert into DB
for team_id, team_data in team_ids_in_matches.items():
    run("""
        INSERT INTO teams (team_id, team_name, teamsname, country)
        VALUES (%s, %s, %s, %s)
        ON CONFLICT (team_id) DO UPDATE
        SET team_name = EXCLUDED.team_name,
            teamsname = EXCLUDED.teamsname,
            country = EXCLUDED.country
        """,
        (team_id, team_data["name"], team_data["sName"], team_data["country"])
    )

# ----------------------------
# 3️⃣ Players Table (for teams appearing in matches)
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS Players(
    player_id INTEGER PRIMARY KEY,
    team_id INTEGER REFERENCES Teams(team_id),
    full_name TEXT,
    batting_style TEXT,
    bowling_style TEXT,
    role TEXT
)
""")

players_list = []
team_ids_in_matches = set()
for m in matches_list:
    if m["team1id"]:
        team_ids_in_matches.add(m["team1id"])
    if m["team2id"]:
        team_ids_in_matches.add(m["team2id"])

for team_id in team_ids_in_matches:
    url = f"https://cricbuzz-cricket.p.rapidapi.com/teams/v1/{team_id}/players"
    try:
        resp = requests.get(url, headers=HEADERS, timeout=10)
        resp.raise_for_status()
        data = resp.json()
    except (requests.RequestException, ValueError) as e:
        print(f"Skipping team {team_id} due to error: {e}")
        continue

    current_role = None
    for player in data.get("player", []):
        if "id" not in player:
            current_role = player.get("name")
            continue
        players_list.append({
            "id": player.get("id"),
            "team_id": team_id,
            "name": player.get("name"),
            "batting": player.get("battingStyle"),
            "bowling": player.get("bowlingStyle"),
            "role": current_role
        })

for p in players_list:
    run("""
    INSERT INTO Players (player_id, team_id, full_name, batting_style, bowling_style, role)
    VALUES (%s, %s, %s, %s, %s, %s)
    ON CONFLICT (player_id) DO UPDATE
    SET team_id = EXCLUDED.team_id,
        full_name = EXCLUDED.full_name,
        batting_style = EXCLUDED.batting_style,
        bowling_style = EXCLUDED.bowling_style,
        role = EXCLUDED.role
    """, (p["id"], p["team_id"], p["name"], p["batting"], p["bowling"], p["role"]))

# ----------------------------
# Venues Table
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS Venues (
    venue_id INTEGER PRIMARY KEY,
    ground TEXT,
    city TEXT,
    country TEXT,
    capacity INTEGER
)
""")

for m in matches_list:
    venue_id = m["venue_id"]
    if venue_id:
        url = f"https://cricbuzz-cricket.p.rapidapi.com/venues/v1/{venue_id}"
        try:
            response = requests.get(url, headers=HEADERS, timeout=10)
            data2 = response.json()
        except Exception as e:
            print(f"Venue {venue_id} fetch error: {e}")
            continue

        api_capacity = str(data2.get("capacity") or "0")
        capacity = int(api_capacity.replace(",", "") or 0)

        run("""
        INSERT INTO Venues (venue_id, ground, city, country, capacity)
        VALUES (%s, %s, %s, %s, %s)
        ON CONFLICT (venue_id) DO UPDATE
        SET ground = EXCLUDED.ground,
            city = EXCLUDED.city,
            country = EXCLUDED.country,
            capacity = EXCLUDED.capacity
        """, (
            data2.get("id"),
            data2.get("ground"),
            data2.get("city"),
            data2.get("country"),
            capacity
        ))

# ----------------------------
# BattingScore & BowlingScore
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS BattingScore (
    player_id INTEGER,
    match_id INTEGER,
    team_id INTEGER,
    name TEXT,
    runs INTEGER,
    balls INTEGER,
    fours INTEGER,
    sixes INTEGER,
    strike_rate REAL,
    PRIMARY KEY(player_id, match_id)
)
""")

run("""
CREATE TABLE IF NOT EXISTS BowlingScore (
    player_id INTEGER,
    match_id INTEGER,
    team_id INTEGER,
    name TEXT,
    overs REAL,
    maidens INTEGER,
    runs INTEGER,
    wickets INTEGER,
    economy REAL,
    PRIMARY KEY(player_id, match_id)
)
""")

# get match ids from DB
cur.execute("SELECT match_id FROM Matches")
match_ids = [row[0] for row in cur.fetchall()]

batting_list = []
bowling_list = []

for MATCH_ID in match_ids:
    scorecard_url = f"https://cricbuzz-cricket.p.rapidapi.com/mcenter/v1/{MATCH_ID}/scard"
    try:
        response = requests.get(scorecard_url, headers=HEADERS, timeout=10)
        scorecard_data = response.json()
    except Exception as e:
        print(f"Skipping scorecard for {MATCH_ID}: {e}")
        continue

    for inning in scorecard_data.get("scorecard", []):
        for batsman in inning.get("batsman", []):
            batting_list.append((
                batsman.get("id"),
                MATCH_ID,
                inning.get("batteamid"),
                batsman.get("name"),
                batsman.get("runs"),
                batsman.get("balls"),
                batsman.get("fours"),
                batsman.get("sixes"),
                float(batsman.get("strkrate") or 0)
            ))
        for bowler in inning.get("bowler", []):
            bowling_list.append((
                bowler.get("id"),
                MATCH_ID,
                inning.get("bowlteamid"),
                bowler.get("name"),
                bowler.get("overs"),
                bowler.get("maidens"),
                bowler.get("runs"),
                bowler.get("wickets"),
                float(bowler.get("econ") or 0)
            ))

# Upsert batting scores
for row in batting_list:
    run("""
    INSERT INTO BattingScore (player_id, match_id, team_id, name, runs, balls, fours, sixes, strike_rate)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    ON CONFLICT (player_id, match_id) DO UPDATE
    SET team_id = EXCLUDED.team_id,
        name = EXCLUDED.name,
        runs = EXCLUDED.runs,
        balls = EXCLUDED.balls,
        fours = EXCLUDED.fours,
        sixes = EXCLUDED.sixes,
        strike_rate = EXCLUDED.strike_rate
    """, row)

# Upsert bowling scores
for row in bowling_list:
    run("""
    INSERT INTO BowlingScore (player_id, match_id, team_id, name, overs, maidens, runs, wickets, economy)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    ON CONFLICT (player_id, match_id) DO UPDATE
    SET team_id = EXCLUDED.team_id,
        name = EXCLUDED.name,
        overs = EXCLUDED.overs,
        maidens = EXCLUDED.maidens,
        runs = EXCLUDED.runs,
        wickets = EXCLUDED.wickets,
        economy = EXCLUDED.economy
    """, row)

# ----------------------------
# FieldingStats
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS FieldingStats (
    player_id INTEGER,
    match_id INTEGER,
    catches INTEGER DEFAULT 0,
    stumpings INTEGER DEFAULT 0,
    PRIMARY KEY(player_id, match_id)
)
""")

fielding_dict = {}
for MATCH_ID in match_ids:
    scorecard_url = f"https://cricbuzz-cricket.p.rapidapi.com/mcenter/v1/{MATCH_ID}/scard"
    try:
        response = requests.get(scorecard_url, headers=HEADERS, timeout=10)
        scorecard_data = response.json()
    except Exception as e:
        print(f"Skipping fielding for {MATCH_ID}: {e}")
        continue

    for team in scorecard_data.get("batTeam", []):
        match_id = scorecard_data.get("matchInfo", {}).get("matchId")
        for batsman in team.get("batsman", []):
            out_desc = batsman.get("outDesc", "") or ""
            fielder_ids = [
                batsman.get("fielderId1", 0),
                batsman.get("fielderId2", 0),
                batsman.get("fielderId3", 0)
            ]
            for f_id in fielder_ids:
                if f_id and f_id != 0:
                    key = (f_id, match_id)
                    if key not in fielding_dict:
                        fielding_dict[key] = {"catches": 0, "stumpings": 0}
                    if "st" in out_desc.lower() or "stumped" in out_desc.lower():
                        fielding_dict[key]["stumpings"] += 1
                    else:
                        fielding_dict[key]["catches"] += 1

for (player_id, match_id), stats in fielding_dict.items():
    run("""
    INSERT INTO FieldingStats (player_id, match_id, catches, stumpings)
    VALUES (%s, %s, %s, %s)
    ON CONFLICT (player_id, match_id) DO UPDATE
    SET catches = EXCLUDED.catches,
        stumpings = EXCLUDED.stumpings
    """, (player_id, match_id, stats["catches"], stats["stumpings"]))

# # ----------------------------
# # Seriesleanback, Matchesleanback, Partnerships, MatchFormatYear
# # ----------------------------

# run("""
# CREATE TABLE IF NOT EXISTS Matchesleanback (
#     match_id INTEGER ,
#     series_id INTEGER,
#     match_format TEXT,
#     status TEXT,
#     toss_winner_id INTEGER,
#     toss_decision TEXT,
#     year INTEGER
# )
# """)
# run("""
# CREATE TABLE IF NOT EXISTS Partnerships (
#     match_id INTEGER,
#     innings_id INTEGER,
#     batsman1_id INTEGER,
#     batsman2_id INTEGER,
#     runs INTEGER
    
# )
# """)

# run("""
# CREATE TABLE IF NOT EXISTS MatchFormatYear (
#     match_id INTEGER ,
#     format TEXT,
#     year INTEGER
# )
# """)

# def safe_int(value):
#     try:
#         return int(value)
#     except (TypeError, ValueError):
#         return None  # if it's "unknown_None" or missing


# for MATCH_ID in match_ids:
#     url = f"https://cricbuzz-cricket.p.rapidapi.com/mcenter/v1/{MATCH_ID}/leanback"
#     try:
#         data = requests.get(url, headers=HEADERS, timeout=10).json()
#     except Exception as e:
#         print(f"Skipping leanback for {MATCH_ID}: {e}")
#         continue

#     match_info = data.get("matchInfo", {})
#     match_id = safe_int(match_info.get("matchId"))
#     series_id = safe_int(match_info.get("seriesId"))
#     toss_winner_id = safe_int(match_info.get("tossResults", {}).get("tossWinnerId"))
#     year = safe_int(match_info.get("year"))
#     match_format = match_info.get("matchFormat")
#     status = match_info.get("status")
#     toss_decision = match_info.get("tossResults", {}).get("decision")
    

#     run("""
#     INSERT INTO Matchesleanback (match_id, series_id, match_format, status, toss_winner_id, toss_decision, year)
#     VALUES (%s, %s, %s, %s, %s, %s, %s)
#     ON CONFLICT (match_id) DO UPDATE
#     SET series_id = EXCLUDED.series_id,
#         match_format = EXCLUDED.match_format,
#         status = EXCLUDED.status,
#         toss_winner_id = EXCLUDED.toss_winner_id,
#         toss_decision = EXCLUDED.toss_decision,
#         year = EXCLUDED.year
#     """, (match_id, series_id, match_format, status, toss_winner_id, toss_decision, year))

#     for inn in data.get("innings", []):
#         innings_id = inn.get("inningsId")
#         for p in inn.get("partnerships", []):
#             run("""
#             INSERT INTO Partnerships (match_id, innings_id, batsman1_id, batsman2_id, runs)
#             VALUES (%s, %s, %s, %s, %s)
#             ON CONFLICT (match_id, innings_id, batsman1_id, batsman2_id) DO UPDATE
#             SET runs = EXCLUDED.runs
#             """, (match_id, innings_id, p.get("batsman1Id"), p.get("batsman2Id"), p.get("runs")))

#     run("""
#     INSERT INTO MatchFormatYear (match_id, format, year)
#     VALUES (%s, %s, %s)
#     ON CONFLICT (match_id) DO UPDATE
#     SET format = EXCLUDED.format,
#         year = EXCLUDED.year
#     """, (match_id, match_format, year))

# ----------------------------
# Recent Matches API (Series table)
# ----------------------------
recent_url = "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/recent"
recent_data = requests.get(recent_url, headers=HEADERS).json()

run("""
CREATE TABLE IF NOT EXISTS Series (
    series_id INTEGER PRIMARY KEY,
    series_name TEXT,
    host_country TEXT,
    match_type TEXT,
    start_date TEXT,
    end_date TEXT,
    total_matches INTEGER
)
""")

def ms_to_date(ms):
    return datetime.utcfromtimestamp(int(ms) / 1000).strftime('%Y-%m-%d') if ms else None

for type_match in recent_data.get("typeMatches", []):
    match_type = type_match.get("matchType")
    for s in type_match.get("seriesMatches", []):
        sw = s.get("seriesAdWrapper", {})
        if not sw:
            continue
        series_id = sw.get("seriesId")
        series_name = sw.get("seriesName")
        total_matches = len(sw.get("matches", []))
        start_date = ms_to_date(sw.get("seriesStartDt"))
        end_date = ms_to_date(sw.get("seriesEndDt"))
        host_country = series_name.split(" tour of ")[-1] if " tour of " in series_name else None

        run("""
        INSERT INTO Series (series_id, series_name, host_country, match_type, start_date, end_date, total_matches)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
        ON CONFLICT (series_id) DO UPDATE
        SET series_name = EXCLUDED.series_name,
            host_country = EXCLUDED.host_country,
            match_type = EXCLUDED.match_type,
            start_date = EXCLUDED.start_date,
            end_date = EXCLUDED.end_date,
            total_matches = EXCLUDED.total_matches
        """, (series_id, series_name, host_country, match_type, start_date, end_date, total_matches))

# ----------------------------
# MostRuns table
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS MostRuns (
    player_id INTEGER PRIMARY KEY,
    player_name TEXT,
    matches INTEGER,
    innings INTEGER,
    runs INTEGER,
    average REAL
)
""")

stats_url = "https://cricbuzz-cricket.p.rapidapi.com/stats/v1/topstats/0"
querystring = {"statsType": "mostRuns", "matchType": "2"}  # test=1, odi=2, t20=3

data = requests.get(stats_url, headers=HEADERS, params=querystring).json()
most_runs_list = []

for item in data.get("values", []):
    values = item.get("values", [])
    if len(values) >= 6:
        most_runs_list.append({
            "player_id": int(values[0]),
            "player_name": values[1],
            "matches": int(values[2]) if values[2] and values[2].isdigit() else None,
            "innings": int(values[3]) if values[3] and values[3].isdigit() else None,
            "runs": int(values[4]) if values[4] and values[4].isdigit() else None,
            "average": float(values[5]) if values[5] else None
        })

for p in most_runs_list:
    run("""
    INSERT INTO MostRuns (player_id, player_name, matches, innings, runs, average)
    VALUES (%s, %s, %s, %s, %s, %s)
    ON CONFLICT (player_id) DO UPDATE
    SET player_name = EXCLUDED.player_name,
        matches = EXCLUDED.matches,
        innings = EXCLUDED.innings,
        runs = EXCLUDED.runs,
        average = EXCLUDED.average
    """, (p["player_id"], p["player_name"], p["matches"], p["innings"], p["runs"], p["average"]))

# cur.execute("""
# INSERT INTO teams (team_id, team_name, teamsname, country)
# VALUES (%s, %s, %s, %s)
# """, (999, "TestTeam", "TT", "TestLand"))
# conn.commit()

# cur.execute("SELECT * FROM teams WHERE team_id = 999;")
# print(cur.fetchall())
 
# ----------------------------
# Close out
# ----------------------------
cur.close()
conn.close()
print("Done — data loaded into PostgreSQL (cricbuzz).")
```

```python
import psycopg2
from psycopg2 import sql
import requests
import time
from datetime import datetime


API_KEY = "997354f96emsh13978d66fe8ee01p1b5c7djsn614af4f1bef5"
API_HOST = "cricbuzz-cricket.p.rapidapi.com"
HEADERS = {"x-rapidapi-key": API_KEY, "x-rapidapi-host": API_HOST}

# ----------------------------
# Connect to PostgreSQL
# ----------------------------
conn = psycopg2.connect(
    dbname="cricbuzz",  # Default database to connect
    user="postgres",
    password="hello01!MSL",
    host="localhost",
    port="5433"
)
cur = conn.cursor()

def run(query, params=None):
    cur.execute(query, params or ())
    
def safe_int(value):
    try:
        return int(value)
    except (TypeError, ValueError):
        return None

def safe_request(url, headers,params=None, max_attempts=3, wait_sec=1):
    """Request with error handling and retry"""
    for attempt in range(max_attempts):
        try:
            resp = requests.get(url, headers=headers, timeout=10)
            resp.raise_for_status()
            return resp.json()
        except Exception as e:
            print(f"Request error for {url}: {e} (attempt {attempt+1})")
            time.sleep(wait_sec)
    return None


# ----------------------------
# 1️⃣ Teams Table
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS Teams (
    team_id INTEGER PRIMARY KEY,
    team_name TEXT,
    teamSName TEXT,
    country TEXT
)
""")
    
# ----------------------------
# 2️⃣ PlayersQuestion1 (single-team example)
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS PlayersQuestion1 (
    player_id INTEGER PRIMARY KEY,
    team_id INTEGER REFERENCES Teams(team_id),
    full_name TEXT,
    batting_style TEXT,
    bowling_style TEXT,
    role TEXT
)
""")

# example: team_id = 2 (you can loop multiple ids if needed)
team_id = 2
url = f"https://cricbuzz-cricket.p.rapidapi.com/teams/v1/{team_id}/players"
data = requests.get(url, headers=HEADERS).json()

playersquestion1 = []
current_role = None
for player in data.get("player", []):
    if "id" not in player:
        current_role = player.get("name")
        continue
    playersquestion1.append({
        "id": player.get("id"),
        "team_id": team_id,
        "name": player.get("name"),
        "batting": player.get("battingStyle"),
        "bowling": player.get("bowlingStyle"),
        "role": current_role
    })

for p in playersquestion1:
    run("""
    INSERT INTO PlayersQuestion1 (player_id, team_id, full_name, batting_style, bowling_style, role)
    VALUES (%s, %s, %s, %s, %s, %s)
    ON CONFLICT (player_id) DO UPDATE
    SET team_id = EXCLUDED.team_id,
        full_name = EXCLUDED.full_name,
        batting_style = EXCLUDED.batting_style,
        bowling_style = EXCLUDED.bowling_style,
        role = EXCLUDED.role
    """, (p["id"], p["team_id"], p["name"], p["batting"], p["bowling"], p["role"]))

# ----------------------------
# 3️⃣ Matches Table
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS Matches (
    match_id INTEGER PRIMARY KEY,
    series_id INTEGER,
    series_name TEXT,
    team1id INTEGER,
    team2id INTEGER,
    team1 TEXT,
    team2 TEXT,
    venue_id INTEGER,
    venuename TEXT,
    city TEXT,
    format TEXT,
    status TEXT,
    state TEXT,
    start_time TEXT,
    result TEXT,
    winner_team_id INTEGER,
    victory_margin TEXT,
    victory_type TEXT,
    toss_winner_id INTEGER,
    toss_decision TEXT
)
""")

match_urls = [
    "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/live",
    "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/recent"
]

matches_list = []
for url in match_urls:
    data = requests.get(url, headers=HEADERS).json()
    for type_match in data.get("typeMatches", []):
        for series in type_match.get("seriesMatches", []):
            series_data = series.get("seriesAdWrapper", {})
            series_name = series_data.get("seriesName")
            series_id = series_data.get("seriesId")
            for match in series_data.get("matches", []):
                info = match.get("matchInfo", {})
                team1 = info.get("team1", {}).get("teamName")
                team2 = info.get("team2", {}).get("teamName")
                team1id = info.get("team1", {}).get("teamId")
                team2id = info.get("team2", {}).get("teamId")
                toss_winner_name = info.get("toss", {}).get("winnerTeam", {}).get("teamName")
                toss_decision = info.get("toss", {}).get("decision")
                toss_winner_id = team1id if toss_winner_name == team1 else (team2id if toss_winner_name == team2 else None)
                status = info.get("status")
                winner_id = None
                victory_margin = None
                victory_type = None
                if status and "won by" in status:
                    parts = status.split(" won by ")
                    winner_name = parts[0].strip()
                    details = parts[1].split(" ")
                    victory_margin = details[0]
                    victory_type = details[1] if len(details) > 1 else ""
                    if winner_name == team1:
                        winner_id = team1id
                    elif winner_name == team2:
                        winner_id = team2id
                matches_list.append({
                    "match_id": info.get("matchId"),
                    "series_id": series_id,
                    "series_name": series_name,
                    "team1id": team1id,
                    "team2id": team2id,
                    "team1": team1,
                    "team2": team2,
                    "venue_id": info.get("venueInfo", {}).get("id"),
                    "venuename": info.get("venueInfo", {}).get("ground"),
                    "city": info.get("venueInfo", {}).get("city"),
                    "format": info.get("matchFormat"),
                    "status": status,
                    "state": info.get("state"),
                    "start_time": info.get("startDate"),
                    "result": info.get("result"),
                    "winner_team_id": winner_id,
                    "victory_margin": victory_margin,
                    "victory_type": victory_type,
                    "toss_winner_id": toss_winner_id,
                    "toss_decision": toss_decision
                })

for m in matches_list:
    run("""
    INSERT INTO Matches
    (match_id, series_id, series_name, team1id, team2id, team1, team2, venue_id, venuename, city, format,
     status, state, start_time, result, winner_team_id, victory_margin, victory_type, toss_winner_id, toss_decision)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    ON CONFLICT (match_id) DO UPDATE
    SET series_id = EXCLUDED.series_id,
        series_name = EXCLUDED.series_name,
        team1id = EXCLUDED.team1id,
        team2id = EXCLUDED.team2id,
        team1 = EXCLUDED.team1,
        team2 = EXCLUDED.team2,
        venue_id = EXCLUDED.venue_id,
        venuename = EXCLUDED.venuename,
        city = EXCLUDED.city,
        format = EXCLUDED.format,
        status = EXCLUDED.status,
        state = EXCLUDED.state,
        start_time = EXCLUDED.start_time,
        result = EXCLUDED.result,
        winner_team_id = EXCLUDED.winner_team_id,
        victory_margin = EXCLUDED.victory_margin,
        victory_type = EXCLUDED.victory_type,
        toss_winner_id = EXCLUDED.toss_winner_id,
        toss_decision = EXCLUDED.toss_decision
    """, (
        m["match_id"], m["series_id"], m["series_name"], m["team1id"], m["team2id"], m["team1"], m["team2"],
        m["venue_id"], m["venuename"], m["city"], m["format"], m["status"], m["state"], m["start_time"],
        m["result"], m["winner_team_id"], m["victory_margin"], m["victory_type"], m["toss_winner_id"], m["toss_decision"]
    ))

#------------------------
#team table 
#------------------------

# ------------------------
# Insert Teams safely
# ------------------------
team_ids_in_matches = {}

for m in matches_list:
    # Add team1
    if m.get("team1id") and m.get("team1name"):
        team_ids_in_matches[m["team1id"]] = {
            "name": m["team1name"],
            "sName": m.get("team1sname", ""),
            "country": None
        }
    # Add team2
    if m.get("team2id") and m.get("team2name"):
        team_ids_in_matches[m["team2id"]] = {
            "name": m["team2name"],
            "sName": m.get("team2sname", ""),
            "country": None
        }

# Insert teams into DB
for team_id, team_data in team_ids_in_matches.items():
    try:
        run("""
            INSERT INTO teams (team_id, team_name, teamsname, country)
            VALUES (%s, %s, %s, %s)
            ON CONFLICT (team_id) DO UPDATE
            SET team_name = EXCLUDED.team_name,
                teamsname = EXCLUDED.teamsname,
                country = EXCLUDED.country
        """, (team_id, team_data["name"], team_data["sName"], team_data["country"]))
    except Exception as e:
        print(f"Error inserting team {team_id}: {e}")

conn.commit()  # Commit after all teams are inserted

# ------------------------
# Insert Players safely
# ------------------------
for player in players_list:  # your fetched players
    team_id = player.get("teamid")
    if not team_id:
        continue  # skip players without a team
    try:
        run("""
            INSERT INTO players (player_id, player_name, team_id, country)
            VALUES (%s, %s, %s, %s)
            ON CONFLICT (player_id) DO UPDATE
            SET player_name = EXCLUDED.player_name,
                team_id = EXCLUDED.team_id,
                country = EXCLUDED.country
        """, (
            player.get("playerid"),
            player.get("playername"),
            team_id,
            player.get("country")
        ))
    except Exception as e:
        print(f"Error inserting player {player.get('playerid')}: {e}")

conn.commit()  # Commit after all players are inserted

  

# ----------------------------
# 3️⃣ Players Table (for teams appearing in matches)
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS Players(
    player_id INTEGER PRIMARY KEY,
    team_id INTEGER ,
    full_name TEXT,
    batting_style TEXT,
    bowling_style TEXT,
    role TEXT
)
""")

players_list = []
team_ids_in_matches = set()
for m in matches_list:
    if m.get("team1id"):
        team_ids_in_matches.add(m["team1id"])
    if m.get("team2id"):
        team_ids_in_matches.add(m["team2id"])

max_requests = 5
for idx, team_id in enumerate(list(team_ids_in_matches)[:max_requests]):
    url = f"https://cricbuzz-cricket.p.rapidapi.com/teams/v1/{team_id}/players"
    data = safe_request(url, HEADERS)
    if not data:
        continue
    current_role = None
    for player in data.get("player", []):
        if "id" not in player:
            current_role = player.get("name")
            continue
        player_id = safe_int(player.get("id"))
        if player_id:
            players_list.append({
                "id": player_id,
                "team_id": team_id,
                "name": player.get("name"),
                "batting": player.get("battingStyle"),
                "bowling": player.get("bowlingStyle"),
                "role": current_role
            })

for p in players_list:
    run("""
    INSERT INTO Players (player_id, team_id, full_name, batting_style, bowling_style, role)
    VALUES (%s, %s, %s, %s, %s, %s)
    ON CONFLICT (player_id) DO UPDATE
    SET team_id = EXCLUDED.team_id,
        full_name = EXCLUDED.full_name,
        batting_style = EXCLUDED.batting_style,
        bowling_style = EXCLUDED.bowling_style,
        role = EXCLUDED.role
    """, (p["id"], p["team_id"], p["name"], p["batting"], p["bowling"], p["role"]))

# ----------------------------
# Venues Table
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS Venues (
    venue_id INTEGER PRIMARY KEY,
    ground TEXT,
    city TEXT,
    country TEXT,
    capacity INTEGER
)
""")

max_requests = 5
for idx, m in enumerate(matches_list[:max_requests]):
    venue_id = m.get("venue_id")
    if not venue_id:
        continue
    
    url = f"https://cricbuzz-cricket.p.rapidapi.com/venues/v1/{venue_id}"
    data2 = safe_request(url, HEADERS)
    if not data2:
        continue

    api_capacity = str(data2.get("capacity") or "0")
    capacity = int(api_capacity.replace(",", "") or 0)

    run("""
    INSERT INTO Venues (venue_id, ground, city, country, capacity)
    VALUES (%s, %s, %s, %s, %s)
    ON CONFLICT (venue_id) DO UPDATE
    SET ground = EXCLUDED.ground,
        city = EXCLUDED.city,
        country = EXCLUDED.country,
        capacity = EXCLUDED.capacity
    """, (
        venue_id,
        data2.get("ground"),
        data2.get("city"),
        data2.get("country"),
        capacity
    ))

# ----------------------------
# BattingScore & BowlingScore
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS BattingScore (
    player_id INTEGER,
    match_id INTEGER,
    team_id INTEGER,
    name TEXT,
    runs INTEGER,
    balls INTEGER,
    fours INTEGER,
    sixes INTEGER,
    strike_rate REAL,
    PRIMARY KEY(player_id, match_id)
)
""")

run("""
CREATE TABLE IF NOT EXISTS BowlingScore (
    player_id INTEGER,
    match_id INTEGER,
    team_id INTEGER,
    name TEXT,
    overs REAL,
    maidens INTEGER,
    runs INTEGER,
    wickets INTEGER,
    economy REAL,
    PRIMARY KEY(player_id, match_id)
)
""")

# get match ids from DB

max_requests = 5

def safe_float(val):
    try:
        return float(val)
    except (TypeError, ValueError):
        return 0.0

batting_list = []
bowling_list = []

for m in matches_list[:max_requests]:
    MATCH_ID = m.get("match_id")
    if not MATCH_ID:
        continue

    scorecard_url = f"https://cricbuzz-cricket.p.rapidapi.com/mcenter/v1/{MATCH_ID}/scard"
    scorecard_data = safe_request(scorecard_url, HEADERS)
    if not scorecard_data:
        continue

    for inning in scorecard_data.get("scorecard", []):
        for batsman in inning.get("batsman", []):
            batting_list.append((
                safe_int(batsman.get("id")),
                MATCH_ID,
                safe_int(inning.get("batteamid")),
                batsman.get("name"),
                safe_int(batsman.get("runs")),
                safe_int(batsman.get("balls")),
                safe_int(batsman.get("fours")),
                safe_int(batsman.get("sixes")),
                safe_float(batsman.get("strkrate"))
            ))
        for bowler in inning.get("bowler", []):
            bowling_list.append((
                safe_int(bowler.get("id")),
                MATCH_ID,
                safe_int(inning.get("bowlteamid")),
                bowler.get("name"),
                safe_float(bowler.get("overs")),
                safe_int(bowler.get("maidens")),
                safe_int(bowler.get("runs")),
                safe_int(bowler.get("wickets")),
                safe_float(bowler.get("econ"))
            ))

# Upsert batting scores
for row in batting_list:
    run("""
    INSERT INTO BattingScore (player_id, match_id, team_id, name, runs, balls, fours, sixes, strike_rate)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    ON CONFLICT (player_id, match_id) DO UPDATE
    SET team_id = EXCLUDED.team_id,
        name = EXCLUDED.name,
        runs = EXCLUDED.runs,
        balls = EXCLUDED.balls,
        fours = EXCLUDED.fours,
        sixes = EXCLUDED.sixes,
        strike_rate = EXCLUDED.strike_rate
    """, row)

# Upsert bowling scores
for row in bowling_list:
    run("""
    INSERT INTO BowlingScore (player_id, match_id, team_id, name, overs, maidens, runs, wickets, economy)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    ON CONFLICT (player_id, match_id) DO UPDATE
    SET team_id = EXCLUDED.team_id,
        name = EXCLUDED.name,
        overs = EXCLUDED.overs,
        maidens = EXCLUDED.maidens,
        runs = EXCLUDED.runs,
        wickets = EXCLUDED.wickets,
        economy = EXCLUDED.economy
    """, row)

# ----------------------------
# FieldingStats
# ----------------------------
run("""
CREATE TABLE IF NOT EXISTS FieldingStats (
    player_id INTEGER,
    match_id INTEGER,
    catches INTEGER DEFAULT 0,
    stumpings INTEGER DEFAULT 0,
    PRIMARY KEY(player_id, match_id)
)
""")
max_requests = 5
fielding_dict = {}

for m in matches_list[:max_requests]:
    MATCH_ID = m.get("match_id")
    if not MATCH_ID:
        continue

    scorecard_url = f"https://cricbuzz-cricket.p.rapidapi.com/mcenter/v1/{MATCH_ID}/scard"
    scorecard_data = safe_request(scorecard_url, HEADERS)
    if not scorecard_data:
        continue

    for team in scorecard_data.get("batTeam", []):
        match_id = safe_int(scorecard_data.get("matchInfo", {}).get("matchId"))
        for batsman in team.get("batsman", []):
            out_desc = batsman.get("outDesc") or ""
            fielder_ids = [
                safe_int(batsman.get("fielderId1")),
                safe_int(batsman.get("fielderId2")),
                safe_int(batsman.get("fielderId3"))
            ]
            for f_id in fielder_ids:
                if f_id and f_id != 0:
                    key = (f_id, match_id)
                    if key not in fielding_dict:
                        fielding_dict[key] = {"catches": 0, "stumpings": 0}
                    if "st" in out_desc.lower() or "stumped" in out_desc.lower():
                        fielding_dict[key]["stumpings"] += 1
                    else:
                        fielding_dict[key]["catches"] += 1

# Upsert fielding stats
for (player_id, match_id), stats in fielding_dict.items():
    run("""
    INSERT INTO FieldingStats (player_id, match_id, catches, stumpings)
    VALUES (%s, %s, %s, %s)
    ON CONFLICT (player_id, match_id) DO UPDATE
    SET catches = EXCLUDED.catches,
        stumpings = EXCLUDED.stumpings
    """, (player_id, match_id, stats["catches"], stats["stumpings"]))

# ----------------------------
# Recent Matches API (Series table)
# ----------------------------

recent_url = "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/recent"
recent_data = safe_request(recent_url, HEADERS)
if not recent_data:
    recent_data = {}

run("""
CREATE TABLE IF NOT EXISTS Series (
    series_id INTEGER PRIMARY KEY,
    series_name TEXT,
    host_country TEXT,
    match_type TEXT,
    start_date TEXT,
    end_date TEXT,
    total_matches INTEGER
)
""")

def ms_to_date(ms):
    return datetime.utcfromtimestamp(int(ms) / 1000).strftime('%Y-%m-%d') if ms else None

for type_match in recent_data.get("typeMatches", []):
    match_type = type_match.get("matchType")
    for s in type_match.get("seriesMatches", []):
        sw = s.get("seriesAdWrapper", {})
        if not sw:
            continue
        series_id = safe_int(sw.get("seriesId"))
        if not series_id:
            continue
        series_name = sw.get("seriesName") or "Unknown"
        total_matches = len(sw.get("matches", []))
        start_date = ms_to_date(sw.get("seriesStartDt"))
        end_date = ms_to_date(sw.get("seriesEndDt"))
        host_country = series_name.split(" tour of ")[-1] if " tour of " in series_name else None

        run("""
        INSERT INTO Series (series_id, series_name, host_country, match_type, start_date, end_date, total_matches)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
        ON CONFLICT (series_id) DO UPDATE
        SET series_name = EXCLUDED.series_name,
            host_country = EXCLUDED.host_country,
            match_type = EXCLUDED.match_type,
            start_date = EXCLUDED.start_date,
            end_date = EXCLUDED.end_date,
            total_matches = EXCLUDED.total_matches
        """, (series_id, series_name, host_country, match_type, start_date, end_date, total_matches))

# ----------------------------
# MostRuns table
# ----------------------------

run("""
CREATE TABLE IF NOT EXISTS MostRuns (
    player_id INTEGER PRIMARY KEY,
    player_name TEXT,
    matches INTEGER,
    innings INTEGER,
    runs INTEGER,
    average REAL
)
""")

stats_url = "https://cricbuzz-cricket.p.rapidapi.com/stats/v1/topstats/0"
querystring = {"statsType": "mostRuns", "matchType": "2"}  # test=1, odi=2, t20=3

data = safe_request(stats_url, HEADERS, params=querystring)
most_runs_list = []

for item in data.get("values", []):
    values = item.get("values", [])
    if len(values) >= 6:
        most_runs_list.append({
            "player_id": safe_int(values[0]),
            "player_name": values[1] or "Unknown",
            "matches": safe_int(values[2]),
            "innings": safe_int(values[3]),
            "runs": safe_int(values[4]),
            "average": safe_float(values[5])
        })

for p in most_runs_list:
    run("""
    INSERT INTO MostRuns (player_id, player_name, matches, innings, runs, average)
    VALUES (%s, %s, %s, %s, %s, %s)
    ON CONFLICT (player_id) DO UPDATE
    SET player_name = EXCLUDED.player_name,
        matches = EXCLUDED.matches,
        innings = EXCLUDED.innings,
        runs = EXCLUDED.runs,
        average = EXCLUDED.average
    """, (p["player_id"], p["player_name"], p["matches"], p["innings"], p["runs"], p["average"]))
# ----------------------------
# Matchesleanback, Partnerships, MatchFormatYear with limit
# ----------------------------

# ---------- TABLE CREATION ----------
run("""
CREATE TABLE IF NOT EXISTS Matchesleanback (
    match_id INTEGER PRIMARY KEY,
    series_id INTEGER,
    match_format TEXT,
    status TEXT,
    toss_winner_id INTEGER,
    toss_decision TEXT,
    year INTEGER
)
""")

run("""
CREATE TABLE IF NOT EXISTS Partnerships (
    match_id INTEGER,
    innings_id INTEGER,
    batsman1_id INTEGER,
    batsman2_id INTEGER,
    runs INTEGER,
    PRIMARY KEY (match_id, innings_id, batsman1_id, batsman2_id)
)
""")

run("""
CREATE TABLE IF NOT EXISTS MatchFormatYear (
    match_id INTEGER PRIMARY KEY,
    format TEXT,
    year INTEGER
)
""")

# ---------- MAIN LOOP WITH LIMIT ----------
max_requests = 5  # set the limit here
for idx, m in enumerate(matches_list[:max_requests]):
    match_id_raw = m.get("match_id")
    if not match_id_raw:
        continue
    
    url = f"https://cricbuzz-cricket.p.rapidapi.com/mcenter/v1/{match_id_raw}/leanback"
    data = safe_request(url, HEADERS)
    if not data:
        continue

    match_info = data.get("matchInfo", {})

    # Extract + sanitize
    match_id = safe_int(match_info.get("matchId"))
    series_id = safe_int(match_info.get("seriesId"))
    toss_winner_id = safe_int(match_info.get("tossResults", {}).get("tossWinnerId"))
    year = safe_int(match_info.get("year"))
    match_format = match_info.get("matchFormat")
    status = match_info.get("status")
    toss_decision = match_info.get("tossResults", {}).get("decision")

    if not match_id:
        print(f"Skipping because match_id missing for raw MATCH_ID={match_id_raw}")
        continue

    # Insert Matchesleanback
    run("""
    INSERT INTO Matchesleanback (match_id, series_id, match_format, status, toss_winner_id, toss_decision, year)
    VALUES (%s, %s, %s, %s, %s, %s, %s)
    ON CONFLICT (match_id) DO UPDATE
    SET series_id = EXCLUDED.series_id,
        match_format = EXCLUDED.match_format,
        status = EXCLUDED.status,
        toss_winner_id = EXCLUDED.toss_winner_id,
        toss_decision = EXCLUDED.toss_decision,
        year = EXCLUDED.year
    """, (match_id, series_id, match_format, status, toss_winner_id, toss_decision, year))

    # Insert Partnerships
    for inn in data.get("innings", []):
        innings_id = safe_int(inn.get("inningsId"))
        for p in inn.get("partnerships", []):
            batsman1_id = safe_int(p.get("batsman1Id"))
            batsman2_id = safe_int(p.get("batsman2Id"))
            runs = safe_int(p.get("runs"))

            if not innings_id or not batsman1_id or not batsman2_id:
                continue

            run("""
            INSERT INTO Partnerships (match_id, innings_id, batsman1_id, batsman2_id, runs)
            VALUES (%s, %s, %s, %s, %s)
            ON CONFLICT (match_id, innings_id, batsman1_id, batsman2_id) DO UPDATE
            SET runs = EXCLUDED.runs
            """, (match_id, innings_id, batsman1_id, batsman2_id, runs))

    # Insert MatchFormatYear
    run("""
    INSERT INTO MatchFormatYear (match_id, format, year)
    VALUES (%s, %s, %s)
    ON CONFLICT (match_id) DO UPDATE
    SET format = EXCLUDED.format,
        year = EXCLUDED.year
    """, (match_id, match_format, year))
# ----------------------------
# Close out
# ----------------------------
cur.close()
conn.close()

print("Done — data loaded into PostgreSQL (cricbuzz).")
```
